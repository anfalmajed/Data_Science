```python
from scipy import stats
grouped_ds = ds.groupby('Hours Studied').mean().reset_index()
x = grouped_ds['Hours Studied']
y = grouped_ds['Performance Index']
slope, intercept, r_value, p_value, std_err = stats.linregress(x, y)
def myfunc(x):
    return slope * x + intercept
mymodel = myfunc(x)
plt.figure(figsize=(6, 5))
plt.scatter(x, y, label='Data Points', color='blue')
plt.plot(x, mymodel, color='red', label='Regression Line')
plt.ylim(ymin=0) 
plt.xlim(xmin=0) 
plt.xlabel('Hours Studied')
plt.ylabel('Performance Index')
plt.title('Regression Line for Hours Studied vs Performance Index')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from scipy import stats
grouped_ds = ds.groupby('Sleep Hours').mean().reset_index()
x = grouped_ds['Sleep Hours']
y = grouped_ds['Performance Index']
slope, intercept, r_value, p_value, std_err = stats.linregress(x, y)
def myfunc(x):
    return slope * x + intercept
mymodel = myfunc(x)
plt.figure(figsize=(7,6))
plt.scatter(x, y, label='Data Points', color='green')
plt.plot(x, mymodel, color='yellow', label='Regression Line')
plt.ylim(ymin=0) 
plt.xlim(xmin=0) 
plt.xlabel('Sleep Hours')
plt.ylabel('Performance Index')
plt.title('Regression Line for Sleep Hours vs Performance Index')
plt.legend()
plt.grid(True)
plt.show()
```
